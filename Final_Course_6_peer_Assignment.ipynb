```python
import sqlite3

# Connect to SQLite database (or create if it doesn't exist)
conn = sqlite3.connect("Course6.db")

# Create a cursor object
cursor = conn.cursor()

#Execute a query
cursor.execute("SELECT * from Census_Data")

# Fetch all rows
rows = cursor.fetchall()

# Print results
for row in rows:
    print(row)


# Close the connection
conn.close()
```

```python
#List all case numbers for crimes involving minors?
%sql SELECT * FROM CRIME_DATA Where lower(Description) like '%minor%'
```

```python
#List all kidnapping crimes involving a child?(children are not considered minors for the purposes of crime analysis)
%sql SELECT * FROM CRIME_DATA Where Primary_Type = 'KIDNAPPING' and Description like '%child%'
```

```python
#What kind of crimes were recorded at schools?
%sql SELECT distinct(Primary_Type) FROM CRIME_DATA where  Location_Description like '%school%'
```

```python
#List the average safety score for all types of schools.
%sql SELECT AVG(SAFETY_SCORE) AS AVERAGE \
FROM PUBLIC_SCHOOLS
```

```python
#List 5 community areas with highest % of households below poverty line.
%sql Select community_Area_Name,Percent_households_below_poverty  \
from Census_Data \
order by Percent_households_below_poverty desc LIMIT 5
```

```python
#Which community area(number) is most crime prone?

%sql Select C.Community_area_Number,Community_area_name, COUNT(CC.COMMUNITY_AREA_NUMBER) AS FREQUENCY  \
from Census_Data C,crime_data CC where C.Community_Area_number=CC.Community_area_Number\
Group BY CC.COMMUNITY_AREA_NUMBER\
ORDER BY COUNT(CC.COMMUNITY_AREA_NUMBER) DESC\
LIMIT 1


```

```python
#Use a sub-query to find the name of the community area with highest hardship index.
%sql SELECT COMMUNITY_AREA_NAME\
FROM  CENSUS_DATA\
WHERE HARDSHIP_INDEX = (SELECT MAX(HARDSHIP_INDEX) FROM CENSUS_DATA);
```

```python
#Problem 10: Use a sub-query to determine the Community Area Name with most number of crimes?
%sql SELECT community_area_name\
FROM CENSUS_DATA \
WHERE COMMUNITY_AREA_NUMBER = (  SELECT CCD.COMMUNITY_AREA_NUMBER \
    FROM CRIME_DATA AS CCD \
    GROUP BY CCD.COMMUNITY_AREA_NUMBER\
    ORDER BY COUNT(CCD.COMMUNITY_AREA_NUMBER) DESC\
    LIMIT 1)\
LIMIT 1;
```
